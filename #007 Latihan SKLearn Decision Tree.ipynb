```
from sklearn.tree import DecisionTreeClassifier
import pandas as pd #Mengimport Library
from sklearn.datasets import load_iris
iris = pd.read_csv('Iris.csv')  #Mengconvert DataSet ke DataFrame
iris.head() #Melihat Informasi Mengenai Data
```

```
iris.drop('Id', axis=1, inplace=True) #Menghapus Kolom Id
x = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]  #Memisahkan Atribut dan Tabel
y = iris['Species']
tree_model = DecisionTreeClassifier() #Membuat Model DecisionTree
tree_model.fit(x, y)  #Melakukan Pelatihan Model Terhadap Data
tree_model.predict([[6.2, 3.4, 5.4, 2.3]])  #Mencoba Model Untuk Prediksi Spesies Bunga Iris
```

```
from sklearn.tree import export_graphviz #Visualisasi dari DecisionTree yang telah di buat
export_graphviz(
    tree_model,
    out_file = "iris_tree.dot",
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-versivirginica'],
    rounded = True,
    filled =True
)
```
